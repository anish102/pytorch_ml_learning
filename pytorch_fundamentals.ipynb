```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print(torch.__version__)
```

```python
# scalar
scalar = torch.tensor(7)
scalar, scalar.ndim, scalar.item()
```

```python
# vector
vector = torch.tensor([7,7])
vector, vector.ndim, vector.shape
```

```python
# matrix
MATRIX = torch.tensor([[7,8],[9,10]])
MATRIX, MATRIX.ndim, MATRIX.shape, MATRIX[1]
```

```python
# tensor
TENSOR = torch.tensor([
    [
        [
            [  12,   -3,  587],
            [  44, 1024,   19],
            [  -7,   18,   91]
        ],
        [
            [  99,   51,  -42],
            [   3,  -17, 2048],
            [ 111,    0,  -13]
        ],
    ],
    [
        [
            [ -88,  233,   10],
            [ 512,  -64,   77],
            [  29,   31, -999]
        ],
        [
            [ 1_000,    5,   -6],
            [   73,  888,   12],
            [  -50,   14,  256]
        ]
    ]
])
TENSOR, TENSOR.ndim, TENSOR.shape
```

```python
# random tensors
random_tensor = torch.rand(1,2,3,4)
random_tensor, random_tensor.ndim, random_tensor.shape
```

```python
# random tensor similar to image tensors
random_image_size_tensor = torch.rand(size=(3,224,224))
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
# tensors with zeros
zeros = torch.zeros(size=(3,4))
zeros, zeros.dtype
```

```python
# tensors with ones
ones = torch.ones(size=(3,4))
ones, ones.dtype
```
